### Data frames
A data frame is more general than a matrix, where different columns can contain different modes of data (numeric, character, etc). It is the same as SAS, SPSS and Stata.

```r
mydata <- data.frame(col1, col2, col3,...)
```

```r
patientID <- c(1, 2, 3, 4)
age <- c(25, 34, 28, 52)
diabetes <- c("Type1","Type2","Type1","Type1")
status <- c("Poor","Improved","Excellent","Poor")
patientdata <- data.frame(patientID, age, diabetes, status)
```

### Lists
Lists are the most complex of the R data type. It allows you to gather a variety of objects under one name.
```r
mylist<-list(object1, object2, object3, ...)
```


```r
g<- "My first list"
h <- c(25, 26, 18, 39)
j <- matrix(1:10, nrow=5)
k <- c("one", "two", "three")
mylist <- list(title=g, ages= h, j, k)
mylist
```

````r
g<- "My first list"
h <- c(25, 26, 18, 39)
j <- matrix(1:10, nrow=5)
k <- c("one", "two", "three")
mylist <- list(title=g, ages= h, j, k)
mylist
````
